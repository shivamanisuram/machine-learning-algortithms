```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer

```

```python
mail_data=pd.read_csv("/content/mail_data.csv")
```

```python

mail_data.head()
```

```python
mail_data.tail()
```

```python
mail_data.shape
```

```python
mail_data.isnull().sum()
```

```python
mail_data=mail_data.dropna()
```

```python
mail_data
```

```python
#replace null values with null string
raw_mail_data = mail_data.where((pd.notnull(mail_data)),'')
```

```python
raw_mail_data.head()
```

```python
raw_mail_data.loc[mail_data['Category']=='spam','Category',]=0
raw_mail_data.loc[mail_data['Category']=='ham','Category',]=1
```

```python
#seperating the data as texts and labels

x=raw_mail_data['Message']
y=raw_mail_data['Category']
```

```python
print(x)
```

```python
print(y)
```

```python
#train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

feature extraction

```python
#transform the text data to feature vectors that can be used as input to logistic regression
```

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True) # Change 'True' to True
x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)
```

```python
print(x_train_features)
```

```python
print(x_test_features)
```

```python
#convert y_train and y_test into integers
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

training machine learning model using logistic regression


```python
model=LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

evaluationg trained model

```python
#predicting training data
prediction_on_training_data=model.predict(x_train_features)
```

```python
print(prediction_on_training_data)
```

```python
accuracy_on_training_data=accuracy_score(y_train,prediction_on_training_data)
```

```python
print(accuracy_on_training_data)
```

```python
prediction_on_test_data=model.predict(x_test_features)
```

```python
print(prediction_on_test_data)
```

```python
prediction_on_test_data=model.predict(x_test_features)
# Calculate accuracy on test data
accuracy_on_test_data=accuracy_score(y_test,prediction_on_test_data) # This line is added
print("accuracy on test data:",accuracy_on_test_data)
```
